```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('D:/Annaconda/data/go_track.csv')
df
```

```python
df = df.drop(["time","rating","rating_bus","rating_weather","car_or_bus","linha"], axis = 1)
df.head()
```

```python
df_x = df.iloc[:, 1:3]
df_x.head()
```

```python
plt.scatter(df.distance, df.speed, s =10, c = "c", marker = "o", alpha = 1)
plt.show()
```

```python
x_array =  np.array(df_x)
print(x_array)
```

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_array)
x_scaled
```

```python
kmeans = KMeans(n_clusters = 2, random_state=123)
kmeans.fit(x_scaled)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
df["kluster"] = kmeans.labels_
```

```python
output = plt.scatter(x_scaled[:,0], x_scaled[:,1], s = 100, c = df.kluster, marker = "o", alpha = 1, )
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='blue', s=200, alpha=1 , marker="s");
plt.title("Hasil Klustering K-Means")
plt.colorbar (output)
plt.show()
```
